```python
from collections import Counter
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource
import pandas

output_notebook()
```

```python
plot = figure()
```

```python
def get_year(date_str):
    return date_str.split('/').pop()

crashes_df = pandas.read_csv('./airplane-crashes.csv')
years_df = crashes_df['Date'].apply(get_year)
crashes_per_year = Counter(years_df)
```

```python
plot.circle(
    list(crashes_per_year.keys()),
    list(crashes_per_year.values()),
    size=5,
    color='navy',
    alpha=0.5
)

show(plot)
```
